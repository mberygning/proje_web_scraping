**Etape 3** *Définition de la fonction fetch_page(url) qui Récupère le contenu HTML de l'URL spécifiée*

```python
def fetch_page(url):
    """
    Récupère le contenu HTML de l'URL spécifiée.

    Args:
    url (str): L'URL du site web à télécharger.

    Returns:
    str: Le contenu HTML de la page.
    """
    response = requests.get(url)
    response.raise_for_status()
    return response.text
```

**Etape 5** *Definition de la fonction  telecharger_et_enregistrer_html(url) qui permet de  télécharger et d'enregistrer le contenu HTML de l'URL spécifiée*

```python
def telecharger_et_enregistrer_html(url):
    """
    Télécharge et enregistre le contenu HTML de l'URL spécifiée.

    Args:
    url (str): L'URL du site web à télécharger.

    Returns:
    None
    """
    try:
        # Utiliser fetch_page pour récupérer le contenu HTML
        contenu_html = fetch_page(url)
        
        # Extraire le nom de domaine de l'URL pour créer un nom de fichier
        url_parsee = urlparse(url)
        domaine = url_parsee.netloc
        nom_fichier = f"{domaine}.html"

        # Enregistrer le contenu HTML dans un fichier
        with open(nom_fichier, 'w', encoding="utf-8") as fichier:
            fichier.write(contenu_html)
        
        print(f"Le contenu HTML a été enregistré dans le fichier: {nom_fichier}")

    except requests.exceptions.RequestException as e:
        print(f"Une erreur s'est produite lors de la récupération de l'URL: {e}")
```

**Etape 7:** *Definition d'une fonction  extract_links_and_texts(soup) pour extraire les differents liens*

```python
# Fonction pour extraire les liens et les textes des balises <li>
def extract_links_and_texts(soup):
    """ 
    Cette fonction extrait les liens (href) et les textes des éléments <a> à l'intérieur des balises <li>
    ayant une classe spécifique. Les informations extraites sont retournées sous forme de liste de dictionnaires,
    où chaque dictionnaire contient le texte du lien et l'URL du lien.
    """
    news_li = []
    # Rechercher les éléments <li> avec une classe spécifique
    for li in soup.find_all('li', class_='header-nav__inner__item'):
        a_tag = li.find('a', href=True)
        if a_tag:
            href = a_tag['href']
            text = a_tag.text.strip()
            # Créer un dictionnaire pour chaque élément
            news_i = {
                'text': text,
                'href': href
            }
            news_li.append(news_i)
    return news_li
```

**Etape 8:** *Definition d'une fonction main(base_url) pour scraper une page web afin d'extraire les liens et les textes des éléments de liste (li)*
    *ayant une classe spécifique. Elle convertit les données extraites en un DataFrame Pandas et les enregistre dans un fichier CSV.*
    *Cette fonction est conçue pour être exécutée avec une URL de base qui contient les éléments de liste ciblés*

```python
# Fonction principale pour scraper les liens et les textes et les mettre dans un DataFrame
def main(base_url):
    """
    La fonction main est utilisée pour scraper une page web afin d'extraire les liens et les textes des éléments de liste (<li>)
    ayant une classe spécifique. Elle convertit les données extraites en un DataFrame Pandas et les enregistre dans un fichier CSV.
    Cette fonction est conçue pour être exécutée avec une URL de base qui contient les éléments de liste ciblés.
    """
    # Obtenir le contenu de la page
    page_content = fetch_page(base_url)
    # Analyser le contenu HTML
    soup = BeautifulSoup(page_content, 'html.parser')
    # Extraire les liens et les textes
    news_li = extract_links_and_texts(soup)
    # Convertir les données en DataFrame
    df = pd.DataFrame(news_li)
    # Afficher le DataFrame
    print(df)
    # Enregistrer le DataFrame dans un fichier CSV
    df.to_csv('links_and_texts.csv', index=False)

# URL de la page à scraper
base_url = "https://www.expat-dakar.com"

# Appeler la fonction principale
main(base_url)
```

**Etape 10:** *Definition d'une fonction extract_vehicle_info(soup, base_url) qui permet d'extraire les informations des véhicules*

```python
# Fonction pour extraire les informations des véhicules
def extract_vehicle_info(soup, base_url):
    """
    La fonction extract_vehicle_info extrait les informations des véhicules à partir d'un objet BeautifulSoup représentant le contenu HTML
    d'une page web.
    Elle recueille des données spécifiques telles que le titre, le prix, l'emplacement, les coordonnées de contact, 
    la date et le lien de chaque véhicule répertorié. 
    Les informations extraites sont ensuite organisées en une liste.
    """

    vehicles = []
    for vehicle_div in soup.find_all('a', class_='listing-card__inner'):
        title_tag = vehicle_div.find('div', class_='listing-card__header__title')
        price_tag = vehicle_div.find('span', class_='listing-card__price__value')
        location_tag = vehicle_div.find('div', class_='listing-card__header__location')
        contact_div = vehicle_div.find('div', class_='listing-card__info-bar__contact')
        date_div=vehicle_div.find('div',class_='listing-card__header__date')
        link_tag = vehicle_div.get('href')  # Extraire le lien
        
        # Assurer que le lien est complet
        link = urljoin(base_url, link_tag) if link_tag else 'N/A'
        
        contact_numbers = []
        if contact_div:
            contact_links = contact_div.find_all('a', class_='listing-card__contact__number-inner')
            for link in contact_links:
                contact_numbers.append(link.text.strip())
                
        contact = ', '.join(contact_numbers)
        
        if title_tag and price_tag and location_tag and date_div :
            title = title_tag.text.strip()
            price = price_tag.text.strip()
            location = location_tag.text.strip().replace('\n', ' ')
            date = date_div.text.strip()
            
            vehicles.append({
                'Title': title,
                'Price': price,
                'Location': location,
                'Contact': contact,
                'Date':date,
                'Link': link
            })
    return vehicles
```

**Etape 11:** *Definition d'une fonction main_vehicules(base_url, start_page, end_page) conçue pour scraper les informations des véhicules à partir d'une série de pages web.* 
     *Elle parcourt les pages spécifiées, extrait les détails des véhicules en utilisant une fonction dédiée, puis compile les données*
     *dans un DataFrame Pandas. Les données sont affichées à l'écran et également  sauvegardées dans un fichier CSV*

```python
# Fonction principale pour scraper les véhicules de pages spécifiques
def main_vehicules(base_url, start_page, end_page):
    """
     La fonction main_vehicules est conçue pour scraper les informations des véhicules à partir d'une série de pages web. 
     Elle parcourt les pages spécifiées, extrait les détails des véhicules en utilisant une fonction dédiée, puis compile les données
     dans un DataFrame Pandas. Les données sont affichées à l'écran et également  sauvegardées dans un fichier CSV
    """
    all_vehicles = []
    
    for page_num in range(start_page, end_page + 1):
        url = f"{base_url}?page={page_num}"
        print(f"Scraping {url}...")
        # Obtenir le contenu de la page des véhicules
        page_content = fetch_page(url)
        # Analyser le contenu HTML
        soup = BeautifulSoup(page_content, 'html.parser')
        # Extraire les informations des véhicules
        vehicles = extract_vehicle_info(soup, base_url)
        all_vehicles.extend(vehicles)

    # Convertir les données en DataFrame
    df = pd.DataFrame(all_vehicles, columns=['Title', 'Price', 'Location', 'Contact','Date', 'Link'])
    # Réorganiser les colonnes pour que 'Link' soit en 6ème position
    df = df[['Title', 'Price', 'Location', 'Contact','Date', 'Link']]
    # Supprimer la colonne 'Link'
    df = df.drop(columns=['Link'])
    # Ajouter une colonne numérotée
    df.index = df.index + 1
    df.index.name = 'No.'
    # Configurer Pandas pour afficher des colonnes plus larges
    pd.set_option('display.max_colwidth', None)
    pd.set_option('display.width', None)
    # Afficher le DataFrame
    print(df.to_string())
    # Optionnel : Enregistrer le DataFrame dans un fichier CSV sans la colonne 'Link'
    df.to_csv('info_vehicules.csv', index=False)  # index=False pour ne pas inclure la numérotation

# URL de la page des véhicules
base_url = "https://www.expat-dakar.com/voitures"

# Appeler la fonction principale pour les pages 1 à 5
main_vehicules(base_url, 1, 5)
```

**Etape 13:** *Ce script est utilisé pour extraire les titres, les comptes, et les liens des catégories d'annonces immobilières depuis une page web spécifique, et stocker ces informations sous forme de liste de dictionnaires pour un traitement ultérieur.*

```python
# Obtenir le contenu de la page immobilier
page_content1 = fetch_page("https://www.expat-dakar.com/immobilier")
# Analyser le contenu HTML
soup = BeautifulSoup(page_content1, 'html.parser')
# Trouver les sections de catégorie
category_elements = soup.select('.filter__category-list-item--level-2')

# Liste pour stocker les données
categories = []

for element in category_elements:
    title = element.select_one('.filter__category-list-item__title').text.strip()
    count = element.select_one('.filter__category-list-item__count').text.strip()
    href = element.select_one('a')['href']
    
    categories.append({
        'Title': title,
        'Count': count,
        'URL': href
    })
```

**Etape 14:** *Ce code est conçu pour extraire des informations concernant les régions disponibles sur une page web d'annonces immobilières en utilisant la bibliothèque BeautifulSoup pour le traitement HTML. Il récupère les titres, les comptes d'annonces et les URLs des régions listées sur la page.*

```python
# Obtenir le contenu de la page immobilier
page_content1 = fetch_page("https://www.expat-dakar.com/immobilier")
# Analyser le contenu HTML
soup = BeautifulSoup(page_content1, 'html.parser')

# Trouver les sections de région
region_elements = soup.select('#filter-region .filter__list-item')

# Liste pour stocker les données
regions = []

for element in region_elements:
    title = element.select_one('.filter__list-item__title').text.strip()
    count = element.select_one('.filter__list-item__count').text.strip()
    href = element.select_one('a')['href']
    
    regions.append({
        'Title': title,
        'Count': count,
        'URL': href
    })
```

**Etape 25:** *Definition d'une fonction extraire_info_offre_demploi qui permet d'extraire des informations des offres d'emploi*

```python
# Fonction pour extraire les informations des offres d'emploi
def extraire_info_offre_demploi(soup, base_url):
    
    """
   La fonction extraire_info_offre_demploi extrait les informations des annonces des offres d'emploi d'une page web HTML en utilisant BeautifulSoup pour 
    le traitement du HTML. Elle retourne une liste de dictionnaires, chaque dictionnaire contenant des détails sur une annonce des offres d'emploi.
    
    """
    offre_d_emploi = []
    for offre_d_emploi_div in soup.find_all('a', class_='listing-card__inner'):
        title_tag = offre_d_emploi_div.find('div', class_='listing-card__header__title')
        price_tag = offre_d_emploi_div.find('span', class_='listing-card__price__value')
        location_tag = offre_d_emploi_div.find('div', class_='listing-card__header__location')
        date_div=offre_d_emploi_div.find('div',class_='listing-card__header__date')
        link_tag = offre_d_emploi_div.get('href')  # Extraire le lien
        
        # Assurer que le lien est complet
        link = urljoin(base_url, link_tag) if link_tag else 'N/A'
        
        if title_tag and price_tag and location_tag and date_div: 
            title = title_tag.text.strip()
            price = price_tag.text.strip()
            location = location_tag.text.strip().replace('\n', ' ')
            date = date_div.text.strip()
            
            offre_d_emploi.append({
                'Title': title,
                'Price': price,
                'Location': location,
                'Date':date,
                'Link': link
            })
    return offre_d_emploi
```

**Etape 26:** *Definition d'une fonction main_offre_demploi conçue pour scraper les informations des offres d'emploi à partir d'une série de pages web.*

```python
# Fonction principale pour scraper les donnees des offres d'emploi de pages spécifiques
def main_offre_demploi(base_url, start_page, end_page):
    """
    La fonction main_offre_demploi est conçue pour scraper les informations des offres d'emploi à partir d'une série de pages web. 
     Elle parcourt les pages spécifiées, extrait les détails des offres d'emploi en utilisant une fonction dédiée, puis compile les données
     dans un DataFrame Pandas. Les données sont affichées à l'écran et également  sauvegardées dans un fichier CSV
    """
    all_offre_demploi = []
    
    for page_num in range(start_page, end_page + 1):
        url = f"{base_url}?page={page_num}"
        print(f"Scraping {url}...")
        # Obtenir le contenu de la page des véhicules
        page_content = fetch_page(url)
        # Analyser le contenu HTML
        soup = BeautifulSoup(page_content, 'html.parser')
        # Extraire les informations des offres d'emploi
        offre_d_emploi = extraire_info_offre_demploi(soup, base_url)
        all_offre_demploi.extend(offre_d_emploi)

    # Convertir les données en DataFrame
    df = pd.DataFrame(all_offre_demploi, columns=['Title', 'Price', 'Location', 'Date', 'Link'])
    # Réorganiser les colonnes pour que 'Link' soit en 6ème position
    #df = df[['Title', 'Price', 'Location', 'Contact','Date', 'Link']]
    # Supprimer la colonne 'Link'
    df = df.drop(columns=['Link'])
    # Ajouter une colonne numérotée
    df.index = df.index + 1
    df.index.name = 'No.'
   
    # Configurer Pandas pour afficher des colonnes plus larges
    pd.set_option('display.max_colwidth', None)
    pd.set_option('display.width', None)
    # Afficher le DataFrame
    print(df.to_string())
    # Optionnel : Enregistrer le DataFrame dans un fichier CSV sans la colonne 'Link'
    df.to_csv('infos_offre_demploi.csv', index=False)  # index=False pour ne pas inclure la numérotation

# URL de la page des offres d'emploi
base_url = "https://www.expat-dakar.com/emploi"

# Appeler la fonction principale pour les pages 1 à 5
main_offre_demploi(base_url, 1, 5)
```

**Etape 36:** *Ce code effectue plusieurs opérations importantes sur les données :*

*Chargement des données dans un DataFrame.*

*Nettoyage et conversion de la colonne des prix en valeurs numériques.*

*Calcul des statistiques descriptives pour les valeurs des prix.*

*Comptage du nombre d'annonces par localisation.*

```python

# Charger les données dans un DataFrame
data = df.head(15)
df = pd.DataFrame(data)

# Nettoyer et convertir la colonne des prix en valeurs numériques
def clean_price(price):
    if isinstance(price, str):
        if price == "Non défini":
            return np.nan
        if '-' in price:
            return float(price.replace('-', '').replace(' ', '').replace('.', ''))
        if 'Entre' in price:
            parts = price.replace('Entre ', '').replace(' et ', '-').replace(' ', '').split('-')
            return (float(parts[0].replace('.', '')) + float(parts[1].replace('.', ''))) / 2
        try:
            return float(price.replace(' ', '').replace('.', ''))
        except ValueError:
            return np.nan
    return np.nan

df['Price'] = df['Price'].apply(clean_price)

# Statistiques descriptives pour les 15 premières valeurs
print("Statistiques descriptives pour les 15 premières valeurs:")
print(df.head(15).describe())

# Nombre d'annonces par location pour les 15 premières valeurs
print("\nNombre d'annonces par location pour les 15 premières valeurs:")
print(df.head(15)['Location'].value_counts())
```
